```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.01

for n in range(101):
    print(n)
    t = n * dt
    data = np.loadtxt(f"csv_output/data{n}.csv", delimiter=",")
    x = data[:, 0]/1000
    u = data[:, 1]
    plt.plot(x, u)
    ax = plt.gca()
    ax.set_xlabel("x")
    ax.set_ylabel("u")
    ax.set_title(f"t = {t:.2f}")
    plt.savefig(f"pngs/u_{n:03d}.png", bbox_inches="tight")
    plt.close()

#use this command to make movie:
#ffmpeg -framerate 10 -i pngs/u_%03d.png -c:v libx264 -vf "pad=ceil(iw/2)*2:ceil(ih/2)*2" -pix_fmt yuv420p advection.mp4
```
